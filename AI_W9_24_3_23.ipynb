```python
from keras.datasets import mnist
(train_images, train_lable),(test_images,test_lable)=mnist.load_data()
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
model=Sequential()
model.add(Dense(512,activation='relu',input_shape=(28*28,)))
model.add(Dense(10,activation='softmax'))
model.summary()
X=test_images
Y=test_lable
```

```python

train_images=train_images.reshape(60000,28*28)
train_images=train_images.astype('float32')/255
test_imagaes=test_images.reshape(10000,28*28)
test_images=test_images.astype('float32')
```

```python
from keras.utils import to_categorical

train_lable=to_categorical(train_lable,10)
test_lable=to_categorical(test_lable,10)
train_lable.shape
test_lable.shape
```

```python

model.compile(optimizer='rmsprop',loss='categorical_crossentropy', metrics=['accuracy'])
model.fit(train_images,train_lable,epochs=10,batch_size=128)
test_loss, test_acc = model.evaluate(test_images,test_lable)
print('mat mat:',test_loss)
print('do chinh xac:',test_acc)
```
